````python
import numpy as np
t = np.linspace(-6,6,20)
sin_t = np.sin(t)
cos_t = np.cos(t)
````

````python
# groupby: splitting a dataframe on values of categorical variables
groupby_gender = data.groupby('Gender')
groupby_gender.mean()
````

````python
for gender, value in groupby_gender['VIQ']:
  print((gender, value.mean()))
````

````python
groupby_gender.std()
````
