## **3. Estimating Pi using Monte Carlo**

The classic example of Monte Carlo simulation is the estimation of &pi;. We are simulating random points in a square that circumscribes a quadrant of a circle and then estimate  &pi; based on the proportion of points that land inside the quadrant.

```
import numpy as np

# Define the number of points
num_samples = 10000

# Generate random points
x = np.random.uniform(0, 1, num_samples)
y = np.random.uniform(0, 1, num_samples)

# calculate the number of points inside the unit of circle quadrant
inside_circle = (x ** 2 + y ** 2) <= 1
pi_estimate = (np.sum(inside_circle) / num_samples) * 4

print('Estimated Pi:', pi_estimate)
```

````python
import numpy as np

# Define the number of points
num_samples = 1000000000

# Generate random points
x = np.random.uniform(0, 1, num_samples)
y = np.random.uniform(0, 1, num_samples)

# calculate the number of points inside the unit of circle quadrant
inside_circle = (x ** 2 + y ** 2) <= 1
pi_estimate = (np.sum(inside_circle) / num_samples) * 4

print('Estimated Pi:', pi_estimate)
````

**Reasoning:**

* **Area Calculation:** The ratio of the area of a circle to the area of the square that encloses it is pi/4.

* **Random Points:** By randomly generating points within the square, the proportion of points falling inside the circle approximates the ratio of their areas.

* **Estimation:** Multiplying this proportion by 4 gives an estimate of pi.

## **4. Financial Risk Assessment**

One real-world application of Monte Carlo simulations is financial risk assessment. Financial Analysts often use it to model asset prices and predict the probability of different outcomes under uncertainty.

```
import numpy as np

# Assuming we have 252 trading days in a year
num_days = 252
num_simulations = 1000

# Starting price of the asset, daily returns mean and standard deviation

start_price = 100
mean_return = 0.001
std_dev = 0.02

# simulate each day's price
price_paths = np.zeros((num_days + 1, num_simulations))
price_paths[0] = start_price

for t in range(1, num_days+1):
  shock = np.random.normal(mean_return, std_dev, num_simulations)
  price_paths[t] = price_paths[t-1] * (1 + shock)

# plotting
import matplotlib.pyplot as plt
plt.plot(price_paths)
plt.xlabel('Days')
plt.ylabel('Asset Price')
plt.title('Monte Carlo Simulation of Asset Prices')
plt.show()
```

````python
import numpy as np

# Assuming we have 252 trading days in a year
num_days = 252
num_simulations = 1000

# Starting price of the asset, daily returns mean and standard deviation

start_price = 100
mean_return = 0.001
std_dev = 0.02

# simulate each day's price
price_paths = np.zeros((num_days + 1, num_simulations))
price_paths[0] = start_price

for t in range(1, num_days+1):
  stock = np.random.normal(mean_return, std_dev, num_simulations)
  price_paths[t] = price_paths[t-1] * (1 + stock)

# plotting
import matplotlib.pyplot as plt
plt.plot(price_paths)
plt.xlabel('Days')
plt.ylabel('Asset Price')
plt.title('Monte Carlo Simulation of Asset Prices')
plt.show()
````

The above code sample generates 'num_simulations' potential future asset price paths over 'num_days', based on a simplifying assumption of daily returns following a normal distribution.
